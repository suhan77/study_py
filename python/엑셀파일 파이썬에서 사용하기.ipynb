```python
input_map = {}

with open(filename, 'r', encoding='utf-8') as meta_f:
    reader = csv.reader(meta_f)
    for line in reader:
        iname = re.sub('\ufeff', '', line[0])
        input_map[iname] = line[1].strip()
```

# 3. UTF8 BOM 읽기

```python
filename = 'csv_bom.csv'
```

```python
input_map = {}
with open(filename, 'rb') as f:
    content = f.read()
    if content.startswith(codecs.BOM_UTF8):
        bom_size = len(codecs.BOM_UTF8)
        content = content[bom_size:]
    string= content.decode('utf8')
    
    reader = csv.reader(string.splitlines())
    for line in reader:
        iname = re.sub('\ufeff', '', line[0])
        input_map[iname] = line[1].strip()
```

```python
input_map = {}

wb = xlrd.open_workbook(filename2)
ws = wb.sheet_by_index(0)
length = ws.nrows
mp3_names = []

for i in range(length):
    value = ws.cell_value(i,0)
    if type(value) == int:
        mp3_names.append(ws.cell_value(i,0))
        print('숫자')
    elif type(value) == float:
        mp3_names.append(str(int(ws.cell_value(i,0))))
        print('숫자')
    else:
        mp3_names.append(ws.cell_value(i,0))
        print('문자')
        
for idx, mp3_name in enumerate(mp3_names):
    input_map[mp3_name] = ws.cell_value(idx,1)
```

```python
def remove_bom(text):
    text = re.sub(codecs.BOM_UTF8.decode('utf-8'), '', text)
    text = re.sub(codecs.BOM_LE.decode("utf-16"), '', text)
    text = re.sub(codecs.BOM_BE.decode('utf-8'), '', text)
    text = re.sub(codecs.BOM_UTF32_LE.decode('utf-8'), '', text)
    text = re.sub(codecs.BOM_UTF32_BE.decode('utf-8'), '', text)
    return text
```

```python
filename = 'utf_8.csv'
input_map = {}

with open(filename, 'r', encoding='utf-8') as meta_f:
    reader = csv.reader(meta_f)
    for line in reader:
        iname = remove_bom(line[0])
        input_map[iname] = line[1].strip()
```
